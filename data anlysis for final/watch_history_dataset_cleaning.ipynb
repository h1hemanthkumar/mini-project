```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("watch_history.csv")
```

```python
data.head(10)
```

```python
data.shape
```

```python
sns.lineplot(x='device_type', y='watch_duration', data=data)
plt.title("Average Watch Duration by Device Type")
plt.show()
```

```python
data['device_type'].value_counts()
```

```python
import pandas as pd

# List of unwanted device types
unwanted_devices = ['Toaster', 'Shoe Box', 'Washing Machine']

# Replace them with 'Other'
data['device_type'] = data['device_type'].replace(unwanted_devices, 'Other')

# Replace NaN values with 'Other'
data['device_type'] = data['device_type'].fillna('Other')
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='device_type', y='watch_duration', data=data)
plt.title("Average Watch Duration by Device Type (Cleaned)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data['watch_duration'].value_counts()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.to_csv("watch_history_modified.csv",index=False)
```
