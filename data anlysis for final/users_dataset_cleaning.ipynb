# Users Data Set cleaning 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
users=pd.read_csv("users.csv")
```

```python
users.head(10)
```

```python
users.shape
```

```python
users.drop_duplicates(inplace=True)
```

```python
users.shape
```

```python
users['age'].value_counts()
```

```python
users['age'] = users['age'].apply(
    lambda x: users['age'].mean() if x > 80 or x <= 0 else x
)
```

```python
users['age'].value_counts()
```

```python
users['email'].value_counts()
```

```python
users.dropna(subset=['email'], inplace=True)
```

```python
users.head(10)
```

```python
users.drop(columns=['email'], inplace=True)
```

```python
users.head(10)
```

```python
users['country'].value_counts()
```

```python
# Define corrections for known inconsistent names
country_map = {
    'INDi': 'India',
    'uSa': 'United States',
    'ruSSUa': 'Russia',
    'shingePore': 'Singapore',
    'palistan': 'Pakistan',
}

# Apply the mapping
users['country'] = users['country'].replace(country_map)
```

```python
users['country'].value_counts()
```

```python
users.to_csv("user_modified.csv",index=True)
```
